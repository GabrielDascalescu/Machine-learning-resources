## Matplotlib example workflow
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# 0. import matplotlib and get it ready for plotting in Jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig , ax = plt.subplots(figsize = (10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize data
ax.set(title = "Simple Plot",
    xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (save the whole figure)
fig.savefig("images/sample-plot.png")


```

```python
x = [1, 2, 4, 8, 16 , 32]
y = [10, 20, 30, 40, 50, 60]

fig , ax = plt.subplots( figsize= (5, 5))

ax.plot(x,y)

ax.set(title = "Lol",
       xlabel = "Jungle dif",
       ylabel = "Kills stolen")

fig.savefig("images/lol.png")
```

### Two options for subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols =2, figsize = (10, 5))
ax1.plot(x, x/2);
ax2. scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000))
ax1.set(title = "Grades");
```

```python
# Option 2
fig, ax = plt.subplots(nrows =2,
                      ncols = 2,
                      figsize = (10, 5))
# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].hist(np.random.randn(1000))
ax[1,1].barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

### Which one should you use ? (pyplot vs matplotlib oo method ?)
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind = "scatter",
            x= "age",
            y = "chol",
            c = "target");
```

```python
# OO method combined with pyplot
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target",
    ax = ax)
ax.set_xlim([45, 100])
ax.set_ylim([0, 600])
```

```python
over_50["target"].values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

#plot the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c= over_50["target"]);

#customize
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol");

#add a legend 
ax.legend(*scatter.legend_elements(),title = "Target");

#add a horizontal line
ax.axhline(over_50.chol.mean(), linestyle = "--");
```

```python
over_50.head()
```

```python
#subplot of chol, age, thalach
fig, (ax1,ax2) = plt.subplots(nrows =2 ,
                              ncols = 1,
                              figsize = (10,10),
                             sharex = True)

#add data 
scatter = ax1.scatter( x = over_50.age,
                     y = over_50.chol,
                     c= over_50.target)

ax1.set(title = "Heart Disease and Cholesterol Levels",
       ylabel = "Cholesterol level")

ax1.legend(*scatter.legend_elements(), title = "Target")
ax1.axhline( y = over_50.chol.mean(), linestyle = '--');

scatter = ax2.scatter( x= over_50.age,
                     y = over_50.thalach,
                     c = over_50.target)

ax2.set(title = "Heart Disease and Max Heart Rate",
       xlabel = "Age",
        ylabel = "Max Heart Rate")

ax2.legend(*scatter.legend_elements(), title = "Target")
ax2.axhline(y = over_50.thalach.mean(), linestyle = '--');

#add a title to the figure;
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
#set the style
plt.style.use('bmh')

# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

#plot the data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c= over_50["target"],
                    cmap = 'winter'); # this changes colour scheme

#customize
ax.set(title = "Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol");

#add a legend 
ax.legend(*scatter.legend_elements(),title = "Target");

#add a horizontal line
ax.axhline(over_50.chol.mean(), linestyle = "--");
```

```python
#customizing y and x axis limitations


#subplot of chol, age, thalach
fig, (ax1,ax2) = plt.subplots(nrows =2 ,
                              ncols = 1,
                              figsize = (10,10),
                             sharex = True)

#add data 
scatter = ax1.scatter( x = over_50.age,
                     y = over_50.chol,
                     c= over_50.target,
                     cmap = 'winter')

ax1.set(title = "Heart Disease and Cholesterol Levels",
       ylabel = "Cholesterol level")

ax1.set_xlim([50,80])

ax1.legend(*scatter.legend_elements(), title = "Target")
ax1.axhline( y = over_50.chol.mean(), linestyle = '--',color = 'red');

scatter = ax2.scatter( x= over_50.age,
                     y = over_50.thalach,
                     c = over_50.target,
                     cmap = 'winter')

ax2.set(title = "Heart Disease and Max Heart Rate",
       xlabel = "Age",
        ylabel = "Max Heart Rate")

ax2.set_xlim([50, 80])
ax2.set_ylim([60,200])

ax2.legend(*scatter.legend_elements(), title = "Target")
ax2.axhline(y = over_50.thalach.mean(), linestyle = '--',color = 'red');

#add a title to the figure;
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
